## 1. Create the function that finds us the total estimated cost of path through node n - F(n)
- Create a way to find cost so far to reach node n - G(n) 
- Create a heuristic function to estimate cost from n to goal - H(n)

```python
def gridPrint(state):
    
    dupState = state.copy()
    dupState[dupState.index(-1)]=" "
    print(dupState[0]," ",dupState[1]," ",dupState[2],end="\n")
    print(dupState[3]," ",dupState[4]," ",dupState[5],end="\n")
    print(dupState[6]," ",dupState[7]," ",dupState[8],end="\n\n")
    
    
    
```

```python
def h(state, target):
    
    manHattenDistance = 0
    for i in state:
        ind1, ind2 = state.index(i), target.index(i)
        y_1, x_1 = ind1 % 3, ind1 // 3
        y_2, x_2 = ind2 % 3, ind2 // 3
        manHattenDistance += abs(x_1-x_2) + abs(y_1-y_2)
        
    return manHattenDistance
```

## 2. Create a Search function
### To traverse across the tree using F(n) to select the next node
    1. Make sure it to discard visited sites
    2. Create possible moves function
    3. Create move generator function


```python
def astar(state, target):
    
    g=1 #starting level assumed as 1
    
    states = [state]
    
    visited_states = []
    while len(states):
        
        all_moves = []
        for state in states:
            visited_states.append(state)
            
            print("Level ",g,end="\n")
            gridPrint(state)
            if state==target:
                print('Success')
                return
            all_moves.extend([move for move in possible_moves(state,visited_states) if move not in all_moves])
            
        
        costs_overall = [g+h(state,target) for state in all_moves]
        
        states = []
        if costs_overall:
            minAmongAll = min(costs_overall)
        
            states = [all_moves[i] for i in range(len(costs_overall)) if costs_overall[i]==minAmongAll]
            g+=1
        
        if costs_overall:
            print("Cost after level {0}: {1}".format(g,min(costs_overall)))
        print("\n")
    print("Failed to search")
        
            
            
        
    
```

```python
def possible_moves(state,visited_states): 
    # Find index of empty spot and assign it to b
    b = state.index(-1)
    
    d = []
    
    if b not in [0,1,2]: 
        d.append('u')
    if b not in [6,7,8]: 
        d.append('d')
    if b not in [0,3,6]: 
        d.append('l')
    if b not in [2,5,8]: 
        d.append('r')
        
    pos_moves_it_can = []
    
    for i in d:
        pos_moves_it_can.append(gen(state,i,b))
        
    return [move_it_can for move_it_can in pos_moves_it_can if move_it_can not in visited_states]
```

```python
def gen(state, direction, blank_state):
    temp = state.copy()                              

    if direction=='d':
        temp[blank_state+3],temp[blank_state] = temp[blank_state],temp[blank_state+3]
    
    if direction=='u':
        temp[blank_state-3],temp[blank_state] = temp[blank_state],temp[blank_state-3]
    
    if direction=='l':
        temp[blank_state-1],temp[blank_state] = temp[blank_state],temp[blank_state-1]
    
    if direction=='r':
        temp[blank_state+1],temp[blank_state] = temp[blank_state],temp[blank_state+1]
        
    return temp
```
